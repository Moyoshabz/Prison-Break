### Importing Pandas and converting our data to a dataframe df

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd

columns = data[0]
rows = data[1:]
df = pd.DataFrame(rows, columns=columns)
```

```python
df.columns = ['Date', 'Prison name', 'Country', 'Succeeded', 'Escapee(s)']
```

```python
print(df.head())
```

```python
print(df.info())
```

```python
print(df.columns)
```

```python
df.columns = df.columns.str.strip()
```

```python
print(df['Succeeded'].unique())
```

### In which countries do helicopter prison breaks have a higher chance of success?

```python
success_counts = df[df["Succeeded"] == "Yes"].groupby("Country").size()
total_attempts = df.groupby("Country").size()
success_rate = (success_counts / total_attempts).fillna(0).sort_values(ascending=False)
```

```python
print(success_counts)
```

### Lets visulaize the countries with the higher chance of a successful prison break

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 8))
success_rate.plot(kind='barh', color='skyblue', edgecolor='black')

plt.title("Helicopter Prison Escape Success Rate by Country", fontsize=14)
plt.xlabel("Success Rate", fontsize=12)
plt.ylabel("Country", fontsize=12)

for index, value in enumerate(success_rate):
    plt.text(value, index, f'{value:.2f}', va='center', ha='left', fontsize=10)

plt.show()
```

### How does the number of escapees affect the success?

```python
df['Num_Escapees'] = df['Escapee(s)'].str.split(',').apply(len)
escapee_success = df[df["Succeeded"] == "Yes"].groupby("Num_Escapees").size()
escapee_attempts = df.groupby("Num_Escapees").size()
escapee_success_rate = (escapee_success / escapee_attempts).fillna(0)
escapee_success_rate_sorted = escapee_success_rate.sort_index()
print(escapee_success_rate_sorted)
```

### Lets visualize how the number of escapees in each attempt correlates with the success rate.

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
escapee_success_rate_sorted.plot(kind='line', color='blue', marker='o')
plt.title("Helicopter Prison Escape Success Rate vs Number of Escapees", fontsize=14)
plt.xlabel("Number of Escapees", fontsize=12)
plt.ylabel("Success Rate", fontsize=12)
plt.show()
```

### Which escapees have done it more than once?

```python
repeat_escapees = df["Escapee(s)"].value_counts()[df["Escapee(s)"].value_counts() > 1]
print(repeat_escapees)
```

#### In the analysis, we see that they were 8 escapee whose name were not included. For the purpose of accurate analysis, i have decided to adopt a form of imputation to classify them as "Unknown"

```python
df["Escapee(s)"] = df["Escapee(s)"].replace("—", "Unknown")
```

Lets run the repeat escapee code again

```python
repeat_escapees = df["Escapee(s)"].value_counts()[df["Escapee(s)"].value_counts() > 1]
print(repeat_escapees)
```

### Lets visualize has attempted to break out of prison by helicopter more than once

```python
plt.figure(figsize=(10, 6))
repeat_escapees.plot(kind='bar', color='lightcoral', edgecolor='black')

plt.title("Repeat Helicopter Prison Escape Attempts", fontsize=14)
plt.xlabel("Escapee(s)", fontsize=12)
plt.ylabel("Number of Attempts", fontsize=12)

plt.xticks(rotation=45, ha='right')

plt.show()
```
